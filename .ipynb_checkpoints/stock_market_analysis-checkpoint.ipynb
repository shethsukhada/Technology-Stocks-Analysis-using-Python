# Value at risk using the "Monte Carlo simulation with GBM" method 
# GBM = geometric Brownian motion 

1. will run many trials with random market conditions
2. then will calculate portfolio losses for each trial.
3. we'll use the aggregation of all these simulations to establish how risky the stock is.

Geometric Browninan motion = ΔS=S(μΔt+σϵ√Δt)

        S = the stock price, 
        mu = the expected return ,
        sigma = standard deviation of the returns, 
        t = time, 
        epsilon = the random variable.

Analyzing Google Stock 

```python
# Setting up time horizon
days = 365

# delta
dt = 1/days

#mu (drift) from the expected return data 
mu = rets.mean()['GOOG']
print(mu)

# volatility of the stock from the std() of the average return
sigma = rets.std()['GOOG']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    ''' starting stock price, days of simulation,mu,sigma, and returns simulated price array'''
    
    #  price array
    price = np.zeros(days)
    price[0] = start_price
    # Schok and Drift
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    # Running price array for number of days
    for x in range(1,days):
        
        # Calculate Schock
        shock[x] = np.random.normal(loc=mu * dt, scale=sigma * np.sqrt(dt))
        # Calculate Drift
        drift[x] = mu * dt
        # Calculate Price
        price[x] = price[x-1] + (price[x-1] * (drift[x] + shock[x]))
        
    return price
```

```python
rets.columns
for col in rets.columns:
    print(col)
    print(rets.mean()[col])
    print(rets.std()[col])
```

```python
for col in rets.columns:
    mu = rets.mean()[col]
    sigma =  rets.std()[col]
    start_price = closing_df[col][0]
    plt.figure()
    for run in range(100):
        plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
    plt.xlabel("Days")
    plt.ylabel("Price")  
    title  =  'Monte Carlo Analysis for ' + col
    plt.title(title)
```

```python
runs = 1000
index = runs
# empty matrix to hold the end price data
simulations = pd.DataFrame(0,index = np.arange(runs), columns = rets.columns)
```

Histogram of the end results for a much larger run. 

```python
# Setting the print options of numpy to only display 0-5 points from an array to suppress output
np.set_printoptions(threshold=5)

col = ''
for col in rets.columns:
    mu = rets.mean()[col]
    sigma =  rets.std()[col]
    start_price = closing_df[col][0]
    for run in range(runs):    
        # Setting the simulation data point as the last stock price for that run
        simulations[col][run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1];
```

```python
# Now let's plot the distribution of the end prices
for col in rets.columns:
    plt.figure()
    title = "Distribution of end prices for " + col
    plt.title(title)
    plt.hist(simulations[col],bins=200)
```

```python
# q as the 1% empirical qunatile, this basically means that 99% of the values should fall between here

for col in rets.columns:
    start_price = closing_df[col][0]
    plt.figure()
    title = "Distribution of end prices for " + col
    q = np.percentile(simulations[col], 1)


    plt.hist(simulations[col],bins=200)

    # Starting Price
    plt.figtext(0.6, 0.8, s="Start price: $%.2f" %start_price)
    # Mean ending price
    plt.figtext(0.6, 0.7, "Mean final price: $%.2f" % simulations[col].mean())

    # Variance of the price (within 99% confidence interval)
    plt.figtext(0.6, 0.6, "VaR(0.99): $%.2f" % (start_price - q,))

    # Display 1% quantile
    plt.figtext(0.15, 0.6, "q(0.99): $%.2f" % q)

    # Plot a line at the 1% quantile result
    plt.axvline(x=q, linewidth=4, color='r')

    # Title
    plt.title("Final price distribution for {} Stock after {} days".format(col ,  days), weight='bold');
    
```

Conclusion : For every initial stock purchase one would be putting $40.38 at risk 99% of the time from our Monte Carlo Simulation.
